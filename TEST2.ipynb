# QUESTION 1

```python
import math
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

y0 = 1

def dydt(y,t):
    dydt = (-2*y*t)
    return dydt
start = 0
end = 2
t= np.linspace(start,end,end)


z=sc.integrate.odeint(dydt,y0,t)
print (z)
```

```python
plt.figure()
plt.plot(t,z,)
plt.title('Solution')
plt.xlabel('t value')
plt.ylabel('y(t) value')
```
